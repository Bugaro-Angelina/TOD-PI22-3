5.2 При помощи объединения таблиц и группировок, создайте DataFrame, состоящий из трех столбцов: recipe_id, name, review_count. У рецептов, для которых отсутствуют отзывы, в соответствущем столбце должен быть указан 0. Подтвердите правильность работы вашего кода, выбрав рецепт, не имеющий отзывов, и выведя на экран строку из полученного DataFrame, содержащую информацию об этом отзыве.


```python
table = pd.merge(recipes[['name', 'id']], reviews['recipe_id'], left_on='id', right_on='recipe_id', how='left')
table
```

```python
my_table = pd.merge(reviews.groupby('recipe_id').count()['date'], table, right_on='recipe_id', 
                    left_on='recipe_id', how='outer').drop('id', axis=1)
my_table = my_table.rename(columns={'date': 'review_count'}).drop_duplicates().reset_index(drop=True)
my_table
```

```python
#Подтвердите правильность работы вашего кода, выбрав рецепт,
# не имеющий отзывов, и выведя на экран строку из полученного
# DataFrame, содержащую информацию об этом отзыве.
my_table['review_count'] = my_table['review_count'].fillna(0)
my_table[my_table['review_count'] == 0]
```
